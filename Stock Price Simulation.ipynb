# Stock Price Simulation

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
days = 30 
```

```python
start_price = 100
```

```python
daily_returns = np.random.normal(0.001,0.02,30)
```

```python
price_series = start_price * np.cumprod(1 + daily_returns)
```

Plotting the Graph

```python
plt.plot(price_series)
```
